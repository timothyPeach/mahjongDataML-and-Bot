```python
def mp(hand):
    markPairs = []
    if len(hand) != 0:
        for i in set(hand):
            if hand.count(i) > 1:
                markPairs.append(i)
    return markPairs

#ot = mp(hand)
#print(ot)
```

#### So I wanna sort out my thoughts on defense against a single reach here, as it's the most common thing you have to defend against, especially in lower level play where most players just reach every time when they are closed and tenpai.

#### The safest tiles against them are the tiles they have discarded, and every tile discarded after they reached, but they might not be the safest tiles period, because someone else might be sitting quietly in tenpai as well, and additionally, if you get rid of all your super safe tiles like honors of which you have the last one (100% safe), then what are you going to do when a second guy reaches?  Would have been nice to have 100% safe tiles for that situation.

#### These are tough consdierations, but because, especially in lower level play, there tends to usually be a very mediocre player at the table, being as short-term-looking with your defense as possible tends to be the best, because the mediocre player is liable to just deal in to the reacher's hand anyways, and so you might as well just be safe now, hoping the hand will end soon.

#### With this in mind, I will now outline an order of safest things top to bottom thinking short-term.

#### Oh one last thing though -- we should probably consider what we will do when we are 1 shanten, because it is possible that we have a way of maintaining our hand with minimal or no risk, and we should prioritize that for sure.  I think the easiest way to do this is to see if the move LookingForward wants to make is within the defensive tiles.  With this in mind, we're going to build defense vs one reach in such a way where it exports 2 lists:

#### The first list is "safe tiles", listed from safest to least safe.
#### The second list is "unsafe tiles" listed from safest to least safe.

#### If a tile is in "safe tiles" and it is what Looking Forward wants to do, we do that move.

#### Safe tiles list construction:
1) Last remaining honor tiles. (100% safe)

2) There is one honor tile other than ours. (99% safe)

3) The tile was discarded by another player after the reacher reached (safe vs 2+ players.)
    -follow the order of honors, terminals, 2/8,3/7,4/6,5 -- this will be called the Standard Defense Order (SDO)
    
4) The tile was discarded by the reacher (use SDO.)

5) 1's underneath 4's, 9's above 6's.  These sorts of things. Use SDO. Also consider how many 1's are out.

6) Tiles beneath or above walls of out-of-play or in-hand tiles. (If there are 3 discarded 2s and you have a 1, the 1 is probably safe.

7) Tiles you have 3 or 4 of.  If you guess right, you can discard it like 2 or 3 more times and you're safe for a while.

```python
def ccT(cT):
    '''put a converted tile in here and get a number out.
    Converts 47, the 7 sou, to the 7 it should be, and converts honor tiles to 1.'''
    try:
        if cT < 50:
            cT = int(str(cT)[1])
        else:
            cT = 1
    except:
        pass
    return cT

def distFrom5(lol):
    #idk why I'm making this, this sentence is longer than the function.
    return abs(lol-5)

def playerPondDist5(lisk):
    vals = lisk[:]
    for i in range(len(vals)):
        vals[i] = distFrom5(ccT(vals[i]))
    return np.mean(vals)

def scaredyCat(ltd=[],atd=[],rtd=[]):
    
    ltdTemp = ltd[:]
    atdTemp = atd[:]
    rtdTemp = rtd[:]
    
    ltdAvg = 0
    atdAvg = 0
    rtdAvg = 0
    
    avgList = [ltdAvg,atdAvg,rtdAvg]
    tempList = [ltdTemp,atdTemp,rtdTemp]
    
    for i in range(len(avgList)):
        if len(tempList[i]) == 0:
            avgList[i] = 4.0
        else:
            avgList[i] = playerPondDist5(tempList[i])
    
    ltdPred = scaredyCatv1.predict_proba([[ltdAvg,len(ltdTemp)]])[0][1]
    atdPred = scaredyCatv1.predict_proba([[atdAvg,len(atdTemp)]])[0][1]
    rtdPred = scaredyCatv1.predict_proba([[rtdAvg,len(rtdTemp)]])[0][1]
    
    return ltdPred,atdPred,rtdPred
    
```

```python
def mouseSpaz(x):
    for _ in range(x):
        x = rn.randint(0,x_res)
        y = rn.randint(0,y_res)
        time.sleep(0.01)
        #print(x)
        #print(y)
        mouse.position = (x,y)

def mm(x,y):
    mouse.position = (x,y)

def buildScreenMatrix():
    xDist = x_res/100.
    yDist = y_res/100.
    #
    xList = []
    xtr = 0
    while xtr < x_res:
        xtr += xDist
        xList.append(xtr)
    #print(len(xList))
    for i in range(len(xList)):
        xList[i] = int(xList[i])
    #
    yList = []
    ytr = 0
    while ytr < y_res:
        ytr += yDist
        yList.append(ytr)
    #print(len(yList))
    for i in range(len(yList)):
        yList[i] = int(yList[i])
    #
    return xList,yList
```
